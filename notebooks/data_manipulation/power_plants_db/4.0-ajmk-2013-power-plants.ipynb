```python
import pandas as pd
```

```python
dat = pd.read_csv('~/Documents/PhD/Projects/10. ELECSIM/elecsim/data/raw/power_plants/uk_power_plants/2013-power-stations.csv')

dat.head(50)
```

```python
dat = dat.rename(columns={'Company Name':'Company', 'Station Name':'Name', 'Fuel':'Fuel', 'Installed':'Capacity', 'Year of Location':'Year_and_Location'})
dat = dat[3:]
dat.head()
```

```python
dat.Company = dat.Company.fillna(method="ffill")
dat.head()
```

```python
dat[['Start_date','Location']]=dat['Year_and_Location'].str.split(' ', 1, expand=True)
dat.head()
```

```python
dat = dat.drop(columns=['Location', 'Year_and_Location'])
```

```python
dat.Start_date = dat.Start_date.str.split('/').str.get(0)
```

```python
dat = dat.dropna()
dat.head()
```

```python
dat = dat[~dat.Capacity.str.contains("Installed")]
dat = dat[~dat.Fuel.str.contains("installed")]
dat
```

```python
dat_renamed = dat.copy()
dat_renamed.Fuel = dat_renamed.Fuel.str.lower()
dat_renamed.Fuel = dat_renamed.Fuel.map(lambda x:"Biomass" if "meat" == x
                                                             else "Offshore" if "wind (offshore)" == x
                                                             else "Biomass" if "biomass_poultry_litter" == x
                                                             else "Biomass" if "straw" == x
                                                             else "Recip_diesel" if "diesel" == x
                                                             else "Biomass" if "biomass_wood" == x
                                                             else "Recip_gas" if "gas" == x
                                                             else "PV" if "solar" == x
                                                             else "OCGT" if "gas oil" == x
                                                             else "EfW" if "waste" == x
                                                             else "Hydro_Store" if "pumped storage" == x
                                                             else "Onshore" if "wind" == x
                                                             else "Coal" if "coal/oil" == x
                                                             else "OCGT" if "gas oil" == x
                                                             else "OCGT" if "gas chp" == x
                                                             else "Biomass" if "rapeseed oil" == x
                                                             else "Hydro" if "hydro/" == x
                                                             else "Biomass" if "coal/biomass" == x
                                                             else "Biomass" if "coal/biomass/" == x
                                                             else "OCGT" if "gas/oil" == x
                                                             else x)

dat_renamed = dat_renamed[~dat_renamed['Fuel'].isin(['mines gas', 'light oil', 'oil', 'gas/coal/oil', 'gas oil/kerosene'])]

pd.unique(dat_renamed.Fuel)
```

```python
dat_renamed.to_csv('~/Documents/PhD/Projects/10. ELECSIM/elecsim/data/processed/power_plants/uk_power_plants/plants_2013/uk_plants_2013.csv')
```
