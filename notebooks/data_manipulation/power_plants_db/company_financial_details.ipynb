## Notebook which connects to Companies House API and retrieves company information. eg. Cash owned

```python
import pandas as pd
```

```python
financials = pd.read_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/data/raw/companies/company_financials.csv')
financials = financials.dropna(how='all')
financials
```

```python
financials_no_source = financials.drop(['Source','Unnamed: 7'], axis=1)
financials_no_stirling = financials_no_source.apply(lambda x: x.str.replace('£',''))
```

```python
financials_no_source
```

```python
# Convert k,m,b to 000, 000000, 000000000 respectively

def unit_to_000(column):
    column = (column.replace(r'[kmb]+$', '', regex=True).astype(float) * 
              column.str.extract(r'[\d\.]+([kmb]+)', expand=False).fillna(1).replace(['k','m','b'], [10**3, 10**6, 10**9]).astype(int))
    return column

financials_to_thousand = financials_no_stirling.copy()
financials_to_thousand[['cash_in_bank','total_liabilities','total_assets','turnover','net_assets']] = financials_no_stirling[['cash_in_bank','total_liabilities','total_assets','turnover','net_assets']].apply(lambda x: unit_to_000(x), axis=0)
financials_to_thousand
```

```python
financials_to_thousand.to_csv('/Users/b1017579/Documents/PhD/Projects/10. ELECSIM/data/processed/companies/company_financials.csv')
```
